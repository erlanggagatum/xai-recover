```python
from softcomputing_packages_bu import *
```

```python
dataset = pd.read_csv('../../../dataset/kepco/household_power_consumption_H.csv', header=None, skiprows=1).values[:, 1:].astype('float')
dataset_norm = np.copy(dataset)
for c in range(8):
    dataset_norm[:, c] -= dataset[:, c].min()
    dataset_norm[:, c] /= (dataset[:, c].max()-dataset[:, c].min())
print(dataset_norm.shape)
```

```python
window_size = 10
n_window = dataset_norm.shape[0]-window_size+1
X, Y = [], []
for w in range(n_window-1):
    X.append(dataset_norm[w:w+window_size])
    Y.append(dataset[w+window_size, 0])
X, Y = np.asarray(X), np.asarray(Y)
print(X.shape, Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=0)
print(X_train.shape, X_test.shape)
print(Y_train.shape, Y_test.shape)
```

```python
input_mlp = Input(shape=(X_train.shape[1:]))
H = Reshape((X_train.shape[1]*X_train.shape[2],))(input_mlp)
H = Dense(64, activation='tanh')(H)
H = Dense(64, activation='tanh')(H)
H = Dropout(0.5)(H)
H = Dense(32, activation='tanh')(H)
output_mlp = Dense(1, activation='linear')(H)
model_mlp = Model(input_mlp, output_mlp)
model_mlp.summary()
model_mlp.compile(loss='mse', optimizer='adam')
es = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
history_model_mlp = model_mlp.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=200, batch_size=2048, verbose=2, shuffle=True, callbacks=[es])
```

```python
Y_test_hat = model_mlp.predict(X_test)
vis_length = 30
x = np.arange(len(Y_test_hat))
for i in range(6):
    rand_target_idx = np.random.randint(0, X_test.shape[0]-window_size)
    plt.plot(x[rand_target_idx:rand_target_idx+vis_length], Y_test[rand_target_idx:rand_target_idx+vis_length], label='Ground Truth')
    plt.plot(x[rand_target_idx:rand_target_idx+vis_length], Y_test_hat[rand_target_idx:rand_target_idx+vis_length], linestyle='--', label='Predicted')
    plt.legend(loc='upper left')
    plt.show()
```

```python
input_cnn_lstm = Input(shape=(X_train.shape[1:]))
H = Reshape((X_train.shape[1], X_train.shape[2], 1))(input_cnn_lstm)
H = TimeDistributed(Conv1D(filters=32, kernel_size=(2), padding='same', activation='tanh'))(H)
H = TimeDistributed(MaxPool1D(2))(H)
H = TimeDistributed(Conv1D(filters=32, kernel_size=(2), padding='same', activation='tanh'))(H)
H = TimeDistributed(MaxPool1D(2))(H)
H = TimeDistributed(Flatten())(H)
H = Dropout(0.5)(H)
H = LSTM(64, return_sequences=True)(H)
H = LSTM(64, return_sequences=False)(H)
H = Dropout(0.5)(H)
H = Dense(64, activation='tanh')(H)
H = Dense(64, activation='tanh')(H)
output_cnn_lstm = Dense(1, activation='linear')(H)
model_cnn_lstm = Model(input_cnn_lstm, output_cnn_lstm)
model_cnn_lstm.summary()
model_cnn_lstm.compile(loss='mse', optimizer='adam')
es = EarlyStopping(monitor='val_loss', patience=5)
history_model_cnn_lstm = model_cnn_lstm.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=200, batch_size=4096, verbose=2, shuffle=True, callbacks=[es])
```

```python
Y_test_hat = model_cnn_lstm.predict(X_test)
vis_length = 30
x = np.arange(len(Y_test_hat))
for i in range(6):
    rand_target_idx = np.random.randint(0, X_test.shape[0]-window_size)
    plt.plot(x[rand_target_idx:rand_target_idx+vis_length], Y_test[rand_target_idx:rand_target_idx+vis_length], label='Ground Truth')
    plt.plot(x[rand_target_idx:rand_target_idx+vis_length], Y_test_hat[rand_target_idx:rand_target_idx+vis_length], linestyle='--', label='Predicted')
    plt.legend(loc='upper left')
    plt.show()
```
